```python
#Importing librares
import pandas as pd
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from matplotlib.pyplot import figure
#matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)


```

```python
#reading the data

Data= pd.read_csv(r"C:\Users\ADMIN\Downloads\movies.csv")
    
```

```python
#previewing first 5 rows
Data.head()
```

```python
#creating a copy of the original dataframe
Data1=Data.copy()
```

```python
#checking for missing data using a for loop

for column in Data.columns:
    
    missing_values = Data[column].isnull().sum()
    
    
    if missing_values > 0:
        print(f"Column '{column}' has {missing_values} missing values.")
    else:
        print(f"Column '{column}' has no missing values.")
```

```python
Data.shape
```

```python
Data1
```

```python

# Impute the mean for null values in each column
Data.fillna(Data.mean(), inplace=True)
```

```python
Data.isnull().sum()
```

```python
Data.shape
```

```python
Data.dropna(inplace=True)
```

```python
#We imputed numerical columns.For the remaining columns with character values,we drop them since it is a small number of rows
Data.isnull().sum()
```

```python
#datatypes for the columns
Data.dtypes
```

```python
#change data type of 'budget' and 'gross' to interger
Data["budget"]=Data["budget"].astype('int64')
Data["gross"]=Data["gross"].astype('int64')
```

```python
#confirming the change
Data.dtypes
```

```python
Data.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#viewing the entire dataframe
pd.set_option('display.max_rows',None)
```

```python
#scatter plot with budget vs. gross revenue
plt.scatter(x=Data['budget'],y=Data['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#plot budget vs gross using sns
sns.regplot(x=Data['budget'],y=Data['gross'],data=Data,scatter_kws={"color":"green"},line_kws={"color":"black"})
```

```python
#pearson,kendall,spearman-can be specified in the argument,i.e method=kendall
Data.corr()#pearson iis the default method-this method checks for correlation in numerical variables
```

```python
#viisualizing 
corr_metrics=Data.corr()
sns.heatmap(corr_metrics,annot=True)
plt.title('Correlation metrics of numeric variables')
plt.show()
```

There is a high correlation between budget and gross. Budget and  votes have a high correlation to gross.
